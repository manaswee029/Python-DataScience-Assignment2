```python
import numpy as np
```

```python
# 1D array with integers from 1 to 20
arr = np.arange(1, 21)
print(arr)
```

```python
# Calculate sum, mean, median, and standard deviation
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
```

```python
# Find indices of elements greater than 10
indices = np.where(arr > 10)
print("Indices of elements greater than 10:", indices)
```

## Part 1: 1D Array Operations

I created a 1D array with numbers from 1 to 20 using `np.arange()`. 
Then I calculated the sum, mean, median, and standard deviation using 
NumPy's built-in functions. I also used `np.where()` to find the 
indices of elements greater than 10.

```python
# 2D array of shape 4x4 with numbers 1 to 16
arr2d = np.arange(1, 17).reshape(4, 4)
print(arr2d)
```

```python
# Transpose of the array
transpose = arr2d.T
print("Transpose:\n", transpose)
```

```python
# Row-wise and column-wise sums
row_sum = np.sum(arr2d, axis=1)
col_sum = np.sum(arr2d, axis=0)

print("Row-wise sum:", row_sum)
print("Column-wise sum:", col_sum)
```

```python
# Two 3x3 arrays with random integers between 1 and 20
arr_a = np.random.randint(1, 21, size=(3, 3))
arr_b = np.random.randint(1, 21, size=(3, 3))

print("Array A:\n", arr_a)
print("\nArray B:\n", arr_b)
```

```python
# Element-wise operations
addition = arr_a + arr_b
subtraction = arr_a - arr_b
multiplication = arr_a * arr_b

print("Addition:\n", addition)
print("\nSubtraction:\n", subtraction)
print("\nMultiplication:\n", multiplication)
```

```python
# Dot product of the two arrays
dot_product = np.dot(arr_a, arr_b)
print("Dot Product:\n", dot_product)
```

## Part 3: Operations on Two 3x3 Arrays

I created two 3x3 arrays with random integers between 1 and 20 using 
`np.random.randint()`. Then I performed element-wise addition, 
subtraction, and multiplication using `+`, `-`, and `*`. I also 
computed the dot product using `np.dot()`, which is different from 
element-wise multiplication — it follows matrix multiplication rules.

```python
# Reshape 1D array of size 12 into 3x4
arr_12 = np.arange(1, 13)
reshaped = arr_12.reshape(3, 4)
print("Reshaped array:\n", reshaped)

# Slice the first two rows and last two columns
sliced = reshaped[:2, 2:]
print("\nSliced array (first 2 rows, last 2 columns):\n", sliced)
```

## Part 4: Reshape and Slicing

I created a 1D array of size 12 and reshaped it into a 3x4 2D array 
using `reshape()`. Then I sliced the array to get the first 2 rows 
and last 2 columns using `reshaped[:2, 2:]`. The `:2` selects rows 
0 and 1, and `2:` selects columns from index 2 onwards.
