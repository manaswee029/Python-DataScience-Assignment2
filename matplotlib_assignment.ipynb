```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Line plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

## Part 1: Line Plot

I created a simple line plot using `plt.plot()` with the given x and y 
values. I added a title using `plt.title()`, axis labels using 
`plt.xlabel()` and `plt.ylabel()`, and enabled grid lines using 
`plt.grid(True)` for better readability.

```python
# Bar graph of student marks
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.bar(students, marks, color=['skyblue', 'lightgreen', 'salmon', 'gold'])
plt.title("Marks Scored by Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

## Part 2: Bar Graph

I created a bar graph using `plt.bar()` to show marks scored by each 
student. I customized the colors of each bar individually using a 
list of color names, and added a title and axis labels.

```python
# Pie chart of revenue distribution
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# Highlight the region with highest revenue using explode
explode = [0.1, 0, 0, 0]  # North America has highest revenue (45%)

plt.pie(revenue, labels=regions, autopct='%1.1f%%', explode=explode, startangle=90)
plt.title("Revenue Distribution by Region")
plt.show()
```

## Part 3: Pie Chart

I created a pie chart using `plt.pie()` to show revenue distribution 
across regions. I used `autopct='%1.1f%%'` to display percentages on 
each slice. To highlight North America (highest revenue at 45%), I 
used the `explode` parameter to slightly pull that slice out from the 
rest of

```python
# Histogram of 1000 random integers between 1 and 100
data = np.random.randint(1, 101, size=1000)

plt.hist(data, bins=20, color='purple', edgecolor='black')
plt.title("Frequency Distribution of Random Integers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## Part 4: Histogram

I generated 1000 random integers between 1 and 100 using 
`np.random.randint()`. Then I used `plt.hist()` to create a histogram 
with 20 bins, which groups the data into ranges and shows how many 
values fall into each range. The `edgecolor='black'` makes the bars 
more distinguishable.
