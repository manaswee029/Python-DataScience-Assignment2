```python
import pandas as pd
```

```python
# Creating a DataFrame with employee data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Print first five rows
print(df.head())
```

```python
# Summary statistics of Age and Salary
print(df[['Age', 'Salary']].describe())
```

```python
# Average salary of HR department
hr_avg_salary = df[df['Department'] == 'HR']['Salary'].mean()
print("Average Salary in HR:", hr_avg_salary)
```

## Part 1: Creating and Exploring the DataFrame

I created a DataFrame using a dictionary with Name, Age, Department, 
and Salary columns. Then I used `.head()` to view the first rows, 
`.describe()` to get summary statistics for Age and Salary, and 
filtered the DataFrame to find the average salary specifically 
for the HR department.

```python
# Add a Bonus column = 10% of salary
df['Bonus'] = df['Salary'] * 0.10
print(df)
```

```python
# Filter employees aged between 25 and 30
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
print(filtered_df)
```

## Part 2 & 3: Adding Bonus Column and Filtering

I added a new 'Bonus' column by multiplying the Salary column by 0.10 
(10%). Then I filtered the DataFrame using boolean conditions with `&` 
to find employees whose age is between 25 and 30. Bob and Eve fit 
this criteria.

```python
# Group by Department and calculate average salary
dept_avg = df.groupby('Department')['Salary'].mean()
print(dept_avg)
```

```python
# Sort by Salary in ascending order and save to CSV
sorted_df = df.sort_values(by='Salary', ascending=True)
print(sorted_df)

sorted_df.to_csv('sorted_employees.csv', index=False)
print("\nSaved to sorted_employees.csv")
```

## Part 4 & 5: Grouping and Sorting

I used `groupby('Department')` to group the data by department and 
calculated the average salary for each using `.mean()`. Finally, I 
sorted the entire DataFrame by Salary in ascending order using 
`sort_values()` and saved the result to a new CSV file using 
`to_csv()`.
